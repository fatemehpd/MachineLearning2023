**Problem 10:** Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python
kn2c_students = { "Name" : ["Fatemeh", "Ali", "Afagh", "Yousef", "Soroush", "Atena"],
        "Age"  : [21, 22, 22, 22, 20 ,21],
        "Score" : [98, 90, 76, 82, 89, 67]}
stu_df = pd.DataFrame(kn2c_students)

grade = []
for i, Score in enumerate(stu_df["Score"]):
  if Score>=90:
    grade.insert(i, 'A')
  elif Score>=80:
    grade.insert(i, 'B')
  elif Score>=70:
    grade.insert(i, 'C')
  else:
    grade.insert(i, 'D')


stu_df["Grade"] = grade
display(stu_df)
````

**Problem 11:** You are given a dataset containing monthly sales data for three products (A, B, and C) over a two-year period. Your task is to perform various data analysis tasks using NumPy, Pandas, and Matplotlib.


*   Dataset:

```
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
```
*  Tasks:
1. Calculate and display the total sales for each product over the two-year period.

2. Calculate and display the average monthly sales for each product.

3. Find the month with the highest sales for each product, and display the product and the sales value for that month.

4. Calculate and display the percentage change in sales for each product from January to December in the second year (2022).

5. Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.

6. Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient.


````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)

#the total sales for each product over the two-year period (task 1)
total_each_product = sales_df.sum(numeric_only=True)
#average monthly sales for each product (task 2)
average_each_product = sales_df.mean(numeric_only=True)
#DataFrame for diaplay (task 1 & 2 display)
pruduct_df = pd.concat([total_each_product,average_each_product], keys = ["Total sales", "Average"], axis=1)
print("\033[33m"+"\nthe total sales and average monthly sales for each product")
display(pruduct_df)

#Find the month with the highest sales for each product (task 3)
highest_sales_month = sales_df.loc[sales_df.idxmax(numeric_only=True)]
print("\033[33m"+"\n\nmonth with the highest sales for each product")
display(highest_sales_month)

#the percentage change in sales for each product from January to December in the second year (task 4)
jan_sales = sales_df.loc[12, 'Product A Sales'::]
dec_sales = sales_df.loc[23, 'Product A Sales'::]
change_jan_dec = ((dec_sales-jan_sales)/jan_sales)*100
change_jan_dec = pd.DataFrame(change_jan_dec)
print("\033[33m"+"\n\nthe percentage change in sales for each product from January to December in the second year")
display(change_jan_dec)


#line chart to visualize the monthly sales data for each product (task 5)
plt.plot(sales_df.Month , sales_df['Product A Sales'], label = "Product A Sales", c = 'g' )
plt.plot(sales_df.Month , sales_df['Product B Sales'], label = "Product B Sales", c = 'b' )
plt.plot(sales_df.Month , sales_df['Product C Sales'], label = "Product C Sales", c = 'r' )
plt.legend()
plt.grid()
plt.title("monthly sales data for each product")
plt.xlabel("month")
plt.ylabel("sales")
print("\n\n")
plt.show()

# correlation between Product A and Product B (task 6)
correlation_AB = sales_df[['Product A Sales', 'Product B Sales']].corr()
if correlation_AB.iloc[0,1]>0.5:
  state = "strong"
else:
  state = "weak"

print(f"\n\ncorrelation between Product A and Product B is {correlation_AB.iloc[0,1]:.3f} which indicates there is a {state} correlation between these two")
````
